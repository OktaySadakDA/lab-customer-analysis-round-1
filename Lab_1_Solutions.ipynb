### 9.Perform the data cleaning operations mentioned so far in class
#### 9.1 Delete the column education and the number of open complaints from the dataframe.

```python
data_new = data.drop(['EDUCATION','NUMBER_OF_OPEN_COMPLAINTS'], axis=1)
data_new
```

#### 9.2 Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

```python
data_new['CUSTOMER_LIFETIME_VALUE'] = data_new['CUSTOMER_LIFETIME_VALUE'].replace(np.nan,0)
```

```python
data_new['CUSTOMER_LIFETIME_VALUE'] = data_new['CUSTOMER_LIFETIME_VALUE'].astype(str)
```

```python
data2 = data_new[data_new['CUSTOMER_LIFETIME_VALUE'].str.contains('%')]
data2.head()
```

```python
data2['CUSTOMER_LIFETIME_VALUE'] = data2['CUSTOMER_LIFETIME_VALUE'].str.replace(r'%$','').astype(float)
data2.head()
```

```python
data2['CUSTOMER_LIFETIME_VALUE'] = data2['CUSTOMER_LIFETIME_VALUE'].apply(lambda x: x*100)
```
